```python
# Task 2

import pandas as pd

df = pd.read_csv(r'C:\KATE\survey_results_public.csv',
               low_memory=False)

schema = pd.read_csv(r'C:\KATE\survey_results_schema.csv')
```

```python
schema_questions = set(schema['qname'])

print(f'Питань у схемі: {len(schema_questions)}')
```

```python
df_columns = set(df.columns)

common_cols = schema_questions.intersection(df_columns)

print(f'Колонок у датасеті:   {len(df_columns)}')
print(f'Питань у схемі:       {len(schema_questions)}')
print(f'Спільних колонок:     {len(common_cols)}')
```

```python
complete_df = df[list(common_cols)].dropna()

complete_respondents = len(complete_df)

print(f'Респондентів без пропусків: {complete_respondents}')
```

```python
# Task 7

# 1. Фільтр Python-розробників
python_df = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# 2. Прибираємо пропуски в компенсації
python_df = python_df.dropna(subset=['ConvertedCompYearly', 'Country'])

# 3. Групування по країнах + агрегація
salary_by_country = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

# 4. Перейменування колонок (для краси)
salary_by_country = salary_by_country.rename(columns={
    'mean': 'AverageSalary',
    'median': 'MedianSalary'
})

# 5. Сортування (опціонально, але красиво)
salary_by_country = salary_by_country.sort_values(by='AverageSalary', ascending=False)

# Вивід
salary_by_country.head(10)
```

Висновок:
Рівень компенсації Python-розробників суттєво відрізняється між країнами.
У деяких країнах спостерігається велика різниця між середнім і медіанним значенням,
що свідчить про наявність викидів (дуже високих або дуже низьких зарплат).
Наприклад, у Viet Nam та Nigeria середнє значення значно перевищує медіану,
що означає, що невелика кількість респондентів з високими доходами впливає на середнє.
Медіана є більш надійним показником "типового" доходу.

```python
# Task 8

# Прибираємо пропуски в зарплаті
salary_df = df.dropna(subset=['ConvertedCompYearly'])

# Сортуємо за спаданням зарплати
top_paid = salary_df.sort_values(by='ConvertedCompYearly', ascending=False)

# Беремо топ-5 і дивимось освіту
top5_education = top_paid[['ConvertedCompYearly', 'EdLevel']].head(5)

print(top5_education)
```

Висновок:
Найбільш оплачувані респонденти мають різні рівні освіти,
включаючи Associate, Bachelor’s та Master’s degree.
Це свідчить про те, що високий рівень доходу не обов'язково
прямо залежить від формальної освіти.

```python
# Task 9

df['UsesPython'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

# Групування по віку
age_stats = df.groupby('Age').agg(
    total_respondents=('ResponseId', 'count'),
    python_users=('UsesPython', 'sum')
)

# Рахуємо відсоток
age_stats['PythonPercentage'] = (age_stats['python_users'] / age_stats['total_respondents'] * 100).round(2)

result = age_stats.sort_values('PythonPercentage', ascending=False)[['total_respondents', 'python_users', 'PythonPercentage']]
result['PythonPercentage'] = result['PythonPercentage'].astype(str) + '%'

print(result)
```

Висновок:
Python найпопулярніший серед наймолодших розробників (18–24 роки) — 40%. У всіх інших вікових групах показник стабільно тримається на рівні 36–39%, і лише серед респондентів 65+ трохи нижчий — 31.63%. Це свідчить про те, що Python є популярним інструментом незалежно від віку.

```python
# Task 10

# 1. Прибираємо пропуски в ключових колонках
clean_df = df.dropna(subset=['ConvertedCompYearly', 'RemoteWork', 'Industry'])

# 2. Обчислюємо 75-й перцентиль
q75 = clean_df['ConvertedCompYearly'].quantile(0.75)

# 3. Фільтруємо:
# - висока зарплата
# - remote робота
top_remote = clean_df[
    (clean_df['ConvertedCompYearly'] > q75) &
    (clean_df['RemoteWork'] == 'Remote')
]

# 4. Найпопулярніші індустрії
top_industries = top_remote['Industry'].value_counts()

print(top_industries.head(10))
```

Висновок:
Найбільш поширеною індустрією серед високооплачуваних віддалених працівників
є Software Development, що значно випереджає інші галузі.
Також популярними є Fintech, Healthcare та Internet/IT-сектор.
Це свідчить про те, що технологічні індустрії пропонують найкращі можливості
для віддаленої роботи з високим рівнем компенсації.
